```python
!pip install --upgrade scikit-learn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from  nltk.corpus  import stopwords
from nltk.stem.porter import  PorterStemmer
import re
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
df = pd.read_csv('/content/training.1600000.processed.noemoticon.csv',encoding='latin-1')
df.head()
```

```python
coln_nme =['target','id','date','flag','user','text']
df.columns = coln_nme
df.head()
```

```python
df.isnull().sum()
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def  streaming(content):
  streaming_content = re.sub('[^a-zA-Z]',' ',content)
  streaming_content = streaming_content.lower()
  streaming_content = streaming_content.split()
  streaming_content = [ps.stem(word) for word in streaming_content if not word in stopwords.words('english')]
  streaming_content = ' '.join(streaming_content)
  return streaming_content
```

```python
df['streaming content'] =df['text'].apply(streaming)
```

```python
print(df['target'])
```

```python
df['target'] = df['target'].replace(4, 1)
```

```python
print (df['target'])
```

```python
x = df['streaming content'].values
y = df['target'].values
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y, random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
vector =CountVectorizer()
```

```python
x_train =vector.fit_transform(x_train)
x_test =vector.transform(x_test)
```

```python
print(x_train)
```

```python
print (x_test)
```

```python
model = LogisticRegression(max_iter =1000)
model.fit(x_train,y_train)
```

```python
x_pred = model.predict(x_train)
acc = accuracy_score(x_pred,y_train)
print(acc)
```

```python
x_testpred = model.predict(x_test)
acc = accuracy_score(x_testpred,y_test)
print(acc)
```
